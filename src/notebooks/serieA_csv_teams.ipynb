```python
from atproto import Client
import os
from dotenv import load_dotenv
import pandas as pd
```

```python
load_dotenv()
```

```python
# Obter as credenciais do arquivo .env
username = os.getenv('BSKY_USERNAME')
password = os.getenv('BSKY_PASSWORD')

# Criação do cliente e login
client = Client()
client.login(username, password)
```

```python
# Handle do time que deseja 
handle = 'saopaulofc.net'
last_post = 'NENHUMA POSTAGEM'
profile_type = 'OFICIAL'
```

```python
# CSV
df = pd.read_csv('archives/serieA_teams.csv')
df.head()
```

```python
# Pegar o perfil
profile = client.get_profile(handle)
print(profile)
```

```python
# Exibe as informações do perfil
print("User Profile:")
print(f"Display Name: {profile.display_name}")
print(f"DID: {profile.did}")
print(f"Description: {profile.description}")
print(f"Handle: {profile.handle}")
print(f"Followers Count: {profile.followers_count}")
print(f"Following Count: {profile.follows_count}")
print(f"Posts Count: {profile.posts_count}")
print(f"Avatar URL: {profile.avatar}")
print(f"Banner URL: {profile.banner}")
print(f"Created At: {profile.created_at}")
print(f"Indexed At: {profile.indexed_at}")
print(f"Joined Via Starter Pack: {profile.joined_via_starter_pack}")
print(f"Labels: {profile.labels}")
print(f"Pinned Post: {profile.pinned_post}")

# Verifica se há informações associadas
associated = profile.associated
if associated:
    print(f"Associated Chat Allow Incoming: {getattr(associated.chat, 'allow_incoming', 'No data')}")
    print(f"Associated Feedgens: {getattr(associated, 'feedgens', 'No data')}")
    print(f"Associated Labeler: {getattr(associated, 'labeler', 'No data')}")
    print(f"Associated Lists: {getattr(associated, 'lists', 'No data')}")
    print(f"Associated Starter Packs: {getattr(associated, 'starter_packs', 'No data')}")
else:
    print("No associated profile found.")

# Viewer State
viewer = profile.viewer
if viewer:
    print(f"Viewer Blocked By: {viewer.blocked_by}")
    print(f"Viewer Blocking: {viewer.blocking}")
    print(f"Viewer Muted: {viewer.muted}")
else:
    print("No viewer information available.")
```

```python
profile_data = {
    'Display Name': profile.display_name,
    'DID': profile.did,
    'Description': profile.description,
    'Handle': profile.handle,
    'Followers Count': profile.followers_count,
    'Following Count': profile.follows_count,
    'Posts Count': profile.posts_count,
    'Avatar URL': profile.avatar,
    'Banner URL': profile.banner,
    'Created At': profile.created_at,
    'Indexed At': profile.indexed_at,
    'Joined Via Starter Pack': profile.joined_via_starter_pack,
    'Labels': profile.labels,
    'Pinned Post': profile.pinned_post,
    'Associated Chat Allow Incoming': getattr(profile.associated.chat, 'allow_incoming',
                                              None) if profile.associated and profile.associated.chat else None,
    'Associated Feedgens': getattr(profile.associated, 'feedgens', None) if profile.associated else None,
    'Associated Labeler': getattr(profile.associated, 'labeler', None) if profile.associated else None,
    'Associated Lists': getattr(profile.associated, 'lists', None) if profile.associated else None,
    'Associated Starter Packs': getattr(profile.associated, 'starter_packs', None) if profile.associated else None,
    'Viewer Blocked By': getattr(profile.viewer, 'blocked_by', None) if profile.viewer else None,
    'Viewer Blocking': getattr(profile.viewer, 'blocking', None) if profile.viewer else None,
    'Viewer Muted': getattr(profile.viewer, 'muted', None) if profile.viewer else None,
    'Profile Type': profile_type,
    'Last Post': last_post,
}
```

```python
# Adiciona em um CSV

# Criar um DataFrame com os dados do novo perfil
new_data = pd.DataFrame([profile_data])

# Concatenar os dados antigos com os novos
df = pd.concat([df, new_data], ignore_index=True)

# Salvar os dados no arquivo CSV
df.to_csv('serieA_teams.csv', index=False)
print("Dados adicionados ao arquivo.")

df
```
